## SERRANO

```python
serrano = pd.read_csv('../sbxw/tables/serrano_filteredlysozyme_filteredwaltzoverlap_filteredlt6.csv')
serrano['dataname'] = 'serrano'
serrano['file'] = serrano['dataname'] + '_' + serrano.index.astype(str) + '.fasta.seq.aggr_profile'
serrano
```

```python
pasta_preds_serrano = glob.glob('2504_benchmarking_for_paper/pasta_serrano157/*.fasta.seq.aggr_profile.dat')
print(len(pasta_preds_serrano))
```

```python
pasta_df = pd.DataFrame(pasta_preds_serrano, columns=['file'])
pasta_df['index'] = pasta_df['file'].apply(lambda x: x.split('_')[5].split('.')[0])
pasta_df_sorted = pasta_df.sort_values(by=['index'], key=lambda x: np.argsort(index_natsorted(pasta_df['index']))).reset_index()[['file', 'index']]
pasta_files = list(pasta_df_sorted['file'])
```

```python
from pathlib import Path
```

```python
data = []
for file in pasta_files:
    sequence = []
    with open(file, 'r') as f:
        d = f.readlines()
        for i in d:
            sequence.append(float(i)) 
    data.append([Path(file).stem,max(sequence),np.array(sequence).mean()])

pasta_data = pd.DataFrame(data,columns=['file','score_max','score_mean'])
pasta_data
```

```python
merge = serrano.merge(pasta_data,on='file')
merge
```

```python
roc_auc_score(merge['value_bool'], merge.score_max)
```

```python
roc_auc_score(merge['value_bool'], merge.score_mean)
```
