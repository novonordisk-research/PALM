```python
import pandas as pd
from sklearn.metrics import roc_auc_score

import lacroix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('resdists.csv')
results
```

```python
results.dataset.unique()
```

```python
fig = plt.figure(figsize=(6,4))
ax1 = sns.boxplot(data=results,x='Residue score',y='dataset',hue='Residue type',fliersize=0, palette=[lacroix.colorList('Lemon')[3],lacroix.colorList('PeachPear')[0],lacroix.colorList('Lime')[4]], boxprops=dict(alpha=0.8))
plt.ylabel('')

plt.legend(fontsize=10, loc='center', bbox_to_anchor=(0, 0, 2.4, 1))
ax1.set_yticklabels(['Non-hydrophobic, 10aa (PALM)', 'WaltzDB non-amyloid, 10aa', 'Random, 10aa', 'Non-hydrophobic, 5aa', 'Non-hydrophobic, 20aa'])

fig.savefig('boxplot.png', dpi=600, bbox_inches='tight')
fig.savefig('supp_resbox.png', dpi=600, bbox_inches='tight')
plt.show()
```
