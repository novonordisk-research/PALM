```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import lacroix
```

```python
app = pd.read_csv('/novo/projects/departments/cdd/public/users/ehec/sbxw_files_2/fibrillatioNN_prediction/sbxw/tables/seuma2022_41467_2022_34742_MOESM6_ESM_allvariants_labels.csv')
palm = pd.read_csv('PALM_app.csv')
palm_wt = pd.read_csv('PALM_wt.csv')
t = pd.read_csv('/novo/projects/departments/cdd/public/users/ehec/au24/2504_benchmarking_for_paper/tango_app_mutant.csv')
app['tango'] = t['Tango']
app['waltz'] = t['Waltz']
t_wt_df = app[app['dataset'] == 'WT']
app_df = app[app['dataset'] != 'WT']
```

```python
fig = plt.figure(figsize=(6,4))
sns.scatterplot(y=palm[palm['is_amyloid'] == False]['sequence_predictions_4'], x=palm[palm['is_amyloid'] == False]['nscore3_c'], label='False', c=lacroix.colorList('PeachPear')[1])
sns.scatterplot(y=palm[palm['is_amyloid'] == True]['sequence_predictions_4'], x=palm[palm['is_amyloid'] == True]['nscore3_c'], label='True', c=lacroix.colorList('Lime')[4])
plt.axhline(y=palm_wt['sequence_predictions_4'][0], color='k', linestyle='--',zorder=0, label='wt')
plt.legend(loc='lower right')
plt.ylabel('PALM sequence score')
plt.xlabel('Nucleation score')
plt.show()
```

```python
fig = plt.figure(figsize=(6,4))
sns.scatterplot(y=app_df[app_df['is_amyloid'] == False]['tango'], x=app_df[app_df['is_amyloid'] == False]['nscore3_c'], label='False', c=lacroix.colorList('PeachPear')[1])
sns.scatterplot(y=app_df[app_df['is_amyloid'] == True]['tango'], x=app_df[app_df['is_amyloid'] == True]['nscore3_c'], label='True', c=lacroix.colorList('Lime')[4])
plt.axhline(y=t_wt_df['tango'][0], color='k', linestyle='--',zorder=0, label='wt')
plt.ylabel('TANGO sequence score')
plt.xlabel('Nucleation score')

plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(3.5, 5))  

gs = gridspec.GridSpec(2, 1)
ax1 = fig.add_subplot(gs[0, 0]) 

sns.scatterplot(y=palm[palm['is_amyloid'] == False]['sequence_predictions_4'], x=palm[palm['is_amyloid'] == False]['nscore3_c'], label='False', c=lacroix.colorList('PeachPear')[1], ax=ax1)
sns.scatterplot(y=palm[palm['is_amyloid'] == True]['sequence_predictions_4'], x=palm[palm['is_amyloid'] == True]['nscore3_c'], label='True', c=lacroix.colorList('Lime')[4], ax=ax1)
plt.axhline(y=palm_wt['sequence_predictions_4'][0], color='k', linestyle='--',zorder=0, label='wt')
plt.legend(loc='lower right')
ax1.set_ylabel('PALM sequence score')
ax1.set_xlabel('')
ax1.set_yticks(np.arange(0.88,1,0.02))
ax1.set_ylim([0.88,0.98])
ax1.get_legend().remove()
ax1.set_xticks(np.arange(-8,6,2))
ax1.set_xlim([-7,5])

ax2 = fig.add_subplot(gs[1, 0]) 

sns.scatterplot(y=app_df[app_df['is_amyloid'] == False]['tango'], x=app_df[app_df['is_amyloid'] == False]['nscore3_c'], label='False', c=lacroix.colorList('PeachPear')[1], ax=ax2)
sns.scatterplot(y=app_df[app_df['is_amyloid'] == True]['tango'], x=app_df[app_df['is_amyloid'] == True]['nscore3_c'], label='True', c=lacroix.colorList('Lime')[4], ax=ax2)
plt.axhline(y=t_wt_df['tango'][0], color='k', linestyle='--',zorder=0, label='wt')
plt.ylabel('TANGO sequence score')
plt.xlabel('Nucleation score')
ax2.get_legend().remove()
ax2.set_yticks(np.arange(8,52,8))
ax2.set_ylim([8,48])

ax2.set_xticks(np.arange(-8,6,2))
ax2.set_xlim([-7,5])

#fig.savefig('/novo/projects/departments/cdd/public/users/ehec/au24/plotting_for_paper/2506_figures_final/fig5.png', dpi=600, bbox_inches='tight')
#fig.savefig('/novo/projects/departments/cdd/public/users/ehec/au24/plotting_for_paper/2506_figures_final/fig5.pdf', dpi=600, bbox_inches='tight')

plt.show()
```
