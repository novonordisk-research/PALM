```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score
```

```python
serrano = pd.read_csv('/novo/projects/departments/cdd/public/users/ehec/sbxw_files_2/fibrillatioNN_prediction/sbxw/tables/serrano_filteredlysozyme_filteredwaltzoverlap_filteredlt6.csv')
true_labels = list(serrano['value_bool'])
```

```python
l = []
with open('/novo/projects/departments/cdd/public/users/ehec/au24/2504_benchmarking_for_paper/aggreprot_exec/out/batch-profile-final-max.txt', 'r') as f:
    for line in f.readlines():
        entries = line.split('\t')
        numerics = [float(i) for i in entries[2:]]
        ans = entries[:2] + [max(numerics)]
        l.append(ans)
```

```python
pred_df = pd.DataFrame(l, columns=['number', 'seq', 'score'])
pred_labels = list(pred_df['score'])
```
