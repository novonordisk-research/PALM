```python
mutations = [['K', 169], ['R', 171], ['E', 221], ['H', 220], ['G', 270], ['R', 270], ['P', 268], ['R', 266], ['D', 285], ['E', 283], ['P', 60]]
doubles = [[['R', 'R'], [163, 175]], [['P', 'E'], [217, 224]], [['E', 'E'], [283, 287]]]

seqs = [linb]
letters = [None]
positions = [None]
old_positions = [None]
for mutant in mutations:
    m_linb = list(linb)
    m_linb[mutant[1]] = mutant[0]
    letters.append([mutant[0]])
    positions.append([mutant[1]])
    old_positions.append([linb[mutant[1]]])
    seqs.append(''.join(m_linb))
    
seqs2 = []
letters2 = []
positions2 = []
old_positions2 = []
for mutant in doubles:
    m_linb = list(linb)
    m_linb[mutant[1][0]] = mutant[0][0]
    m_linb[mutant[1][1]] = mutant[0][1]
    letters2.append(mutant[0])
    positions2.append(mutant[1])
    old_positions2.append([linb[mutant[1][0]], linb[mutant[1][1]]])
    seqs2.append(''.join(m_linb))

seqs = seqs + seqs2
letters = letters + letters2
positions = positions + positions2
olds = old_positions + old_positions2

data = pd.DataFrame(columns=['sequence'])
data['sequence'] = seqs
data['new_residue'] = letters
data['mutation_position'] = positions
data['original_residue'] = olds

linbs = data
```
