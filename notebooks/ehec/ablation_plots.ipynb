```python
import pandas as pd
import numpy as np
from statsmodels.stats.multicomp import pairwise_tukeyhsd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ablations = pd.read_csv('250415_crossval_p2.csv')
```

```python
tukey = pairwise_tukeyhsd(endog=ablations["amypro22 rocauc"], groups=ablations["model"], alpha=0.05)
best = ablations.groupby(by='model').mean(numeric_only=True).reset_index().sort_values("amypro22 rocauc",ascending=False)
```

```python
tukey.plot_simultaneous(comparison_name=best['model'].reset_index(drop=True)[0])
plt.title('amypro rocauc')
```
