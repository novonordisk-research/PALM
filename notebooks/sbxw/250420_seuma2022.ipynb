```python
path = 'tables/seuma2022_41467_2022_34742_MOESM6_ESM_allvariants.csv'
df = pd.read_csv(path)
filt_df = df[df['dataset'].isin(['single'])].copy(deep=True)
filt_df['position'] = filt_df['ID'].apply(lambda x: int(x.split('-')[1]))
# 672 is the first residue in amyloid beta peptide (D)
filt_df['mutation_fullseq'] = filt_df['ID'].apply(lambda x: x[0]+str(int(x[2:-2])+671)+x[-1])
filt_df.sort_values(by='position',inplace=True)
print(filt_df.shape)
display(filt_df.groupby('fAD').size())
filt_df[filt_df['fAD']=='fAD_d']
```
