```python
import pandas as pd
import numpy as np
import json
import torch
```

```python
path = 'anupp/250216_ANNuPP_feature_vectors.csv'
df = pd.read_csv(path).set_index('amino_acid')
df
```

```python
anupp_dict = dict()
for aa,row in df.iterrows():
    anupp_dict[aa] = list(row)

with open('anupp/250216_ANNuPP_feature_vectors.json','w') as file:
    json.dump(anupp_dict,file)
```

```python
print(anupp_dict)
```

```python
test_seq = 'RSFFSFLGEAFDGARDMWRAYSDMREANYIGSDKYFHARGNYDAAKRGPGGVWAAEAISDARENIQRFFGHGAEDSLADQAANEWGRSGKDPNHFRPAGLPEKY'
feature_list = []
for char in test_seq:
    feature_list.append(anupp_dict[char])
seq_features = torch.tensor(feature_list,dtype=float)
seq_features
```

```python
sequence_embeddings = torch.stack([x.mean(dim=0) for x in seq_features])
sequence_embeddings.numpy()
```

```python
from abc import ABC, abstractmethod

class AbstractClass(ABC):
    @property
    @abstractmethod
    def my_property(self):
        pass

class DerivedClass(AbstractClass):
    @property
    def my_property(self):
        return "Property value from DerivedClass"

# Usage
obj = DerivedClass()
print(obj.my_property)  # Output: Property value from DerivedClass
```
