```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
nnk4 = pd.read_csv('nnk4_v1.csv')
#nnk4 = nnk4.dropna()
nnk4.head(3)
```

```python
nnk4['sequence'] = nnk4['aa_seq'].apply(lambda x: x.split('*')[0])
```

```python
nnk4['len'] = nnk4['sequence'].apply(lambda x: len(x))
```

```python
plt.scatter(range(len(nnk4)), nnk4['len'])
```

```python
nnk4['value_bool'] = nnk4['nucleator']
t = nnk4[['sequence', 'len', 'value_bool']].dropna()
t.to_csv('nnk4_truefull_nostop.csv')
```
